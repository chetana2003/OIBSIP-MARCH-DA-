```python
df = pd.read_csv("/content/drive/MyDrive/Data Analytics/AB_NYC_2019.csv")
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.violinplot(data = df, x = df['price'])
```

```python
percentiles = np.linspace(.05, 1, num = 20)
df.describe(percentiles = percentiles)
```

```python
percentile_05 = df['price'].quantile(.05)
percentile_95 = df['price'].quantile(.95)

df_clean = df[(percentile_05 <= df['price']) & (df['price'] <= percentile_95)]
```

```python
sns.violinplot(data = df_clean, x = df_clean['price'])
```
