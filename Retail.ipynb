```python
import pandas as pd

rs = pd.read_csv("/content/drive/MyDrive/Data Analytics/retail_sales_dataset (1).csv")
#rs

# delete unnecessary fields(in my case).
rs = rs[['Date', 'Gender', 'Age', 'Product Category', 'Quantity', 'Price per Unit', 'Total Amount']]
rs
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set(style="whitegrid")

# Create a pairplot to visualize the age distribution by gender
plt.figure(figsize=(10, 6))
sns.histplot(data=rs, x='Age', hue='Gender', kde=True, bins=10, common_norm=False)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution by Gender')
```

```python
Product_Category_List = rs['Product Category'].unique()
print(Product_Category_List)
```

```python
male_favorite_category = rs[rs['Gender'] == 'Male']['Product Category'].mode().values[0]
female_favorite_category = rs[rs['Gender'] == 'Female']['Product Category'].mode().values[0]

print(f"[Men] Most Popular Product Category: {male_favorite_category}")
print(f"[Women] Most Popular Product Category: {female_favorite_category}")
```

```python
sns.set(style="whitegrid")

# Create a scatterplot to visualize the correlation between Age and Quantity
plt.figure(figsize=(8, 6))
sns.scatterplot(data=rs, x='Quantity', y='Total Amount', hue='Gender', alpha=0.7)
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.title('Correlation between Quantity and Total Amount')
plt.legend(title='Gender')

plt.show()

# Calculate the correlation coefficient between Age and Quantity
correlation_coefficient = rs['Quantity'].corr(rs['Total Amount'])

print(f"Correlation coefficient between Quantity and Total Amount: {correlation_coefficient}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
menu = pd.read_csv('/content/drive/MyDrive/Data Analytics/menu.csv')
menu.info()
```

```python
plt.figure(figsize=(15,8))
sns.kdeplot(data  = menu , x = 'Calories' , hue = 'Category'  )
plt.show()
```

```python
menu.iloc[menu[menu['Category'] == 'Chicken & Fish']['Calories'].idxmax()]
```

```python
def confidence_interval():
    categories = menu['Category'].unique()
    X_bars  , LCLs , UPCs = [] , [] , []
    for c in categories:
        c_df = menu[menu['Category'] == c]['Calories']
        X_bar , std = np.mean(c_df) , np.std(c_df)
        LCL , UPC = max(X_bar - ( 1.96 * std) , 0) ,  X_bar + ( 1.96 * std)
        LCLs.append(LCL)
        UPCs.append(UPC)
        X_bars.append(X_bar)
    return pd.DataFrame({'Category' : categories , 'LCL' : LCLs ,'X-bar' : X_bars ,'UCL' : UPCs} )
```

```python
confidence_interval()
```
